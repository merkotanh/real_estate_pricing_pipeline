# Model Inspection

This notebook loads the trained model and explores:
- model coefficients
- feature importance (linear)
- interpretation of model behavior

```python
import joblib
import sys
import pandas as pd
from pathlib import Path

current = Path().resolve()

# project root
while current != current.parent:
    if (current / 'src').exists():
        PROJECT_ROOT = current
        break
    current = current.parent

MODEL_DIR = PROJECT_ROOT / 'model'
```

```python
data_rent = joblib.load(MODEL_DIR / 'rent_model_bundle.pkl')
data_buy = joblib.load(MODEL_DIR /  'buy_model_bundle.pkl')

print(data_rent)
print(data_buy)
```

```python
df_info = pd.DataFrame({
    'feature': data_rent['features'],
    'coef': data_rent['model'].coef_
})

df_info
```

```python
df_info = pd.DataFrame({
    'feature': data_buy['features'],
    'coef': data_buy['model'].coef_
})

df_info
```

### Interpretation

- Positive coefficient → increases price per sqm  
- Negative coefficient → decreases price per sqm  

Market saturation shows a negative effect, consistent with supply-demand theory.
